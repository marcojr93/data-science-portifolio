### The Scenario

You are a data analyst working for an ice cream company. Management is interested in improving the company's ice cream sales.

The company has been collecting data about its sales—but not a lot. The available data is from an internal data source and is based on sales for 2019. You’ve been asked to review the data and provide some insight into the company’s ice cream sales. Ideally, management would like answers to the following questions:

What is the most popular flavor of ice cream?

How does temperature affect sales?

Is there a part of the year that sells more ice cream?

#### Import libraries and datasets

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
color = sns.color_palette()
sns.set_style('darkgrid')
import warnings
warnings.filterwarnings(action = 'ignore')
import pandas as pd
```

```python
# Importing all datasets for analysis
df1 = pd.read_excel('Sales_by_day.xlsx')
```

```python
df2 = pd.read_excel('Sales_by_temperature.xlsx')
```

```python
df3 = pd.read_excel('Sales_by_flavour.xlsx')
```

#### What is the most popular flavor of ice cream?

```python
# There is one column with "weeks" at the original dataset which we don't need yet, so we can drop
df4 = df3.drop(['week'], axis =1)
display(df4)
```

```python
# This information can be made simple by analyzing the number of units sold by flavor
dfsoma = df4.groupby(['flavor']).sum()
print(dfsoma)
```

###### So, we can conclude that Lemon, is tem most popular flavor based on its units sold

#### How does temperature affect sales?

```python
# We can use a graphic for analyse this question
ax= sns.lmplot(x="temperature", y="sales", data=df2)
ax.fig.suptitle('Sales x Temperature', fontsize=16, y=1.02)
ax.set_xlabels("Max temperature", fontsize=14)
ax.set_ylabels("Sales", fontsize=14)
ax
     
```

###### It's easy to verify that the Sales increases with temperature, and the range between 60-70º is the best moment to sell ice creams
