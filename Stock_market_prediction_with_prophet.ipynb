```python
import pandas as pd
import yfinance as yf
from datetime import datetime
from datetime import timedelta
import plotly.graph_objects as go
from fbprophet import Prophet
from fbprophet.plot import plot_plotly, plot_components_plotly
import warnings
warnings.filterwarnings('ignore')
pd.options.display.float_format = '${:,.2f}'.format
```

## Setup

```python
today = datetime.today().strftime('%Y-%m-%d')
initial_date = '2019-01-01'
df_stock = yf.download('TSLA', initial_date, today) # For brazilian stocks, it's necessary use ".SA" in the ticker end (eg: ITSA4.SA) 
df_stock.tail()
df_stock.reset_index(inplace = True)
df = df_stock[["Date", "Adj Close"]]
df.rename(columns={'Date':'ds', "Adj Close":"y"}, inplace = True) # Prophet requires two columns, X and Y
fig = go.Figure()
fig.add_trace(go.Scatter(x = df['ds'], y = df['y']))
```

```python
model = Prophet()
model.fit(df)
df_future = model.make_future_dataframe(periods = 300)
prediction = model.predict(df_future)
prediction[['ds', 'yhat', 'yhat_lower', 'yhat_upper']]
plot_plotly(model, prediction)
```

The Tesla stock price target by the end of year it's between US$ 1441 and 1597

```python
plot_components_plotly(model, prediction)
```

The best day to buy TSLA is Friday, that is supose to be the lowest price during the week
