# EDA with python

The main goal of this project is analyse what kind of service it will be need to be offer to increase the number of constumers


## Libraries and dataset

```python
# Importing pandas librarie
import pandas as pd
df = pd.read_csv('telecom_database.csv')
display(df)
```

```python
#The Unnamed Columns is not necessary

df = df.drop(['Unnamed: 0'], axis =1)
display(df)
```

```python
# Some information abaout dataset

print(df.info())
```

## Database cleanning

```python
# The columns TotalCharges is classified as Object, but actually it's a columns of numbers, so we need to transform to numeric

df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
print(df.info())
```

```python
# It is also necessary to delete all columns with empty values

df = df.dropna(how = "all", axis = 1) # To delete columns
df = df.dropna(how = "any", axis = 0) # To delete lines
print(df.info())
```

## EDA

```python
# Churn show if the client still with the company or no, and will halp the company with future decisions based on the other features

display(df['Churn'].value_counts(normalize = True).map("{:.1%}".format))
```

```python
# Plotly is a graph librarie that will help in data visualization
  
import plotly.express as px
column = 'Tenure'
graf = px.histogram(df, x = column, color = 'Churn')
graf.show()
```

This graph show that the time is crucial for the churn. How long the client staty with the company, less the number of churn

### Churn x All columns graphs

```python
# For loop to plot all graphs to analyse the churn pattern

import plotly.express as px

for column in df.columns:
  print(column)
  graph = px.histogram(df, x = column, color = 'Churn')
  graph.show()
```
